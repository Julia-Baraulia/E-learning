## Task 6
## Often, for a qualitative analysis of the audience, I use approaches based on segmentation.

### Using Python, create adapted RFM clusters of students to qualitatively assess your audience.
### In the adapted clustering, you can choose the following metrics: R - average time to pass one exam, F - course completion rate, M - average score obtained in the exam.
### Describe in detail how you created the clusters.
### For each RFM segment, establish boundaries for the recency, frequency, and monetary metrics for interpreting these clusters.

 Calculating Recency - average time to pass one exam.

```python
Recency =score_exam.groupby('id_student')\
    .agg({'date_submitted':'mean'})\
    .sort_values('date_submitted', ascending = False)
Recency
```

 Calculating Frequency - course completion rate.

```python
attempts = score_exam.groupby('id_student')\
    .agg({'date_submitted':'count'})\
    .sort_values('date_submitted', ascending = False)
successfull_attempt = score_40.groupby('id_student')\
    .agg({'date_submitted':'count'})\
    .sort_values('date_submitted', ascending = False)
Frequency = attempts.merge(successfull_attempt, on = 'id_student', how = 'left')
Frequency['success_rate'] = Frequency['date_submitted_y'] / Frequency['date_submitted_x']
Frequency = Frequency[['success_rate']]
Frequency.fillna(0)
Frequency=Frequency.dropna()
Frequency
```

Calculating Monetary - average number of points.

```python
Monetary = score_exam.groupby('id_student')\
    .agg({'score':'mean'})\
    .sort_values('score', ascending = False)
Monetary
```

```python
rfm = Recency.merge(Frequency, on = 'id_student', how = 'outer')
rfm = rfm.merge(Monetary, on = 'id_student')
rfm = rfm.rename(columns={"date_submitted": "Recency", "success_rate": "Frequency", "score":'Monetary'})
rfm
```


Let's calculate the quantiles for each metric.

```python
quintiles = rfm[['Recency', 'Frequency', 'Monetary']].quantile([.2, .4, .6, .8]).to_dict()
quintiles
```

```python
quintiles = rfm[['Recency', 'Frequency', 'Monetary']].quantile([.2, .4, .6, .8]).to_dict()
quintiles
def r_score(x):
    if x <= quintiles['Recency'][.2]:
        return 5
    elif x <= quintiles['Recency'][.4]:
        return 4
    elif x <= quintiles['Recency'][.6]:
        return 3
    elif x <= quintiles['Recency'][.8]:
        return 2
    else:
        return 1

def fm_score(x, c):
    if x <= quintiles[c][.2]:
        return 1
    elif x <= quintiles[c][.4]:
        return 2
    elif x <= quintiles[c][.6]:
        return 3
    elif x <= quintiles[c][.8]:
        return 4
    else:
        return 5
```

```python
rfm['R'] = rfm['Recency'].apply(lambda x: r_score(x))
rfm['F'] = rfm['Frequency'].apply(lambda x: fm_score(x, 'Frequency'))
rfm['M'] = rfm['Monetary'].apply(lambda x: fm_score(x, 'Monetary'))
```

```python
rfm['RFM Score'] = rfm['R'].map(str) + rfm['F'].map(str) + rfm['M'].map(str)
rfm = rfm.sort_values('RFM Score')
rfm
```

```python
 segments = {
    r'[4-5][1-2]': 'zubrilki',
    r'[1-2][4-5]': 'poslednie_ryadi',
    r'[4-5]5': 'vunderkindi',
    r'[4-5][3-4]': 'horoshisti',
    r'35': 'uspeshnie',
    r'3[3-4]': 'srednyachki',
    r'3[1-2]': 'nizhe_srednego',
    r'[1-2][1-2]': 'dvoechniki',
    r'[1-2]3': 'tusovchiki'
}

rfm['Segment'] = rfm['M'].map(str) + rfm['R'].map(str)
rfm['Segment'] = rfm['Segment'].replace(segments, regex=True)
rfm.head()
```

## Here are the 9 segments::
1) zubrilki - spend a lot of time and perform well in exams.;

2) poslednie_ryadi - spend little time and perform poorly in exams;

3) vunderkindi - spend little time and still perform well in exams;

4) horoshisti  - spend enough time and achieve high grades;

5) uspeshnie - spend little time and get average grades;

6) srednyachki - spend an average amount of time and receive average grades;

7) nizhe_srednego - spend a lot of time but achieve not very high grades, perhaps needing more time or assistance;

8) dvoechniki - spend a lot of time but receive low grades (or cram at the last moment without spending much time);

9) tusovchiki - receive low grades and could dedicate more time.
